````python
# 1) Setup
import os, json, re, time, random
import pandas as pd
from openai import OpenAI

os.environ["OPENAI_API_KEY"] = "REDACTED_OPENAI_KEY"
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
MODEL = "gpt-5.4"

def _resp_text(resp):
    if getattr(resp, "output_text", None):
        return resp.output_text
    chunks = []
    for item in getattr(resp, "output", []):
        for c in getattr(item, "content", []):
            t = getattr(c, "text", None)
            if t:
                chunks.append(t)
    return "\n".join(chunks)

def _parse_json(text):
    text = text.strip()
    text = re.sub(r"^```(?:json)?\s*|\s*```$", "", text, flags=re.I | re.S).strip()
    return json.loads(text)

def ask_json(system_prompt, user_payload, retries=3):
    user_text = user_payload if isinstance(user_payload, str) else json.dumps(user_payload, ensure_ascii=False)
    last_err = None

    for i in range(retries):
        try:
            resp = client.chat.completions.create(
                model=MODEL,
                temperature=0,
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": user_text},
                ],
                response_format={"type": "json_object"},
            )
            return json.loads(resp.choices[0].message.content)
        except Exception as e:
            last_err = e
            time.sleep(1.2 * (i + 1))

    raise last_err
````

````python
# 2) Prepare brief-level discovery sample (one brief -> many successful query variants)
import pandas as pd
import re
import random

cols = ["description", "sanitized_search_term", "query", "fq_values"]
x = good_collections[cols].copy()

for c in cols:
    x[c] = x[c].fillna("").astype(str)

# keep only rows with a brief
x = x[x["description"].str.strip() != ""].reset_index(drop=True)

def norm_desc(s: str) -> str:
    s = s.strip().lower()
    s = re.sub(r"\s+", " ", s)
    return s

def _clip(s: str, n: int) -> str:
    return s[:n] if len(s) > n else s

x["brief_id"] = x["description"].map(norm_desc)
x["description_short"] = x["description"].map(lambda s: _clip(s, 700))
x["sanitized_search_term_short"] = x["sanitized_search_term"].map(lambda s: _clip(s, 220))
x["query_short"] = x["query"].map(lambda s: _clip(s, 220))
x["fq_values_short"] = x["fq_values"].map(lambda s: _clip(s, 260))

# signature for de-duping near-identical variants under the same brief
x["variant_sig"] = (
    x["sanitized_search_term_short"].str.strip().str.lower() + " || " +
    x["query_short"].str.strip().str.lower() + " || " +
    x["fq_values_short"].str.strip().str.lower()
)

def pick_variants(g: pd.DataFrame, max_terms: int = 12, seed: int = 42):
    g2 = g.drop_duplicates("variant_sig")
    rows = g2[["sanitized_search_term_short", "query_short", "fq_values_short"]].rename(columns={
        "sanitized_search_term_short": "sanitized_search_term",
        "query_short": "query",
        "fq_values_short": "fq_values",
    }).to_dict("records")
    if len(rows) <= max_terms:
        return rows
    rng = random.Random(seed)
    rng.shuffle(rows)
    return rows[:max_terms]

# aggregate to brief-level objects
brief_objects = (
    x.groupby("brief_id", as_index=False)
     .apply(lambda g: pd.Series({
         "description": g["description_short"].iloc[0],
         "term_variants": pick_variants(g, max_terms=12, seed=42),
         "variant_count_total": int(g["variant_sig"].nunique()),
     }))
     .reset_index(drop=True)
)

brief_objects["brief_row_id"] = brief_objects.index.astype(str)

# optional: brief-level stratification (so discovery sees diverse brief shapes)
brief_objects["variant_bin"] = pd.qcut(
    brief_objects["variant_count_total"].rank(method="first"),
    q=min(4, max(1, brief_objects["variant_count_total"].nunique())),
    labels=False,
    duplicates="drop"
)
brief_objects["desc_len"] = brief_objects["description"].str.len().clip(upper=5000)
brief_objects["desc_bin"] = pd.qcut(
    brief_objects["desc_len"].rank(method="first"),
    q=5,
    labels=False,
    duplicates="drop"
)
brief_objects["strata"] = brief_objects["variant_bin"].astype(str) + "|" + brief_objects["desc_bin"].astype(str)

def stratified_brief_sample(frame: pd.DataFrame, n: int = 240, seed: int = 42):
    n = min(n, len(frame))
    groups = [g for _, g in frame.groupby("strata")]
    per = max(1, n // max(1, len(groups)))
    chunks = [g.sample(min(len(g), per), random_state=seed) for g in groups]
    out = pd.concat(chunks).drop_duplicates(subset="brief_row_id")
    need = n - len(out)
    if need > 0:
        remaining = frame[~frame["brief_row_id"].isin(out["brief_row_id"])]
        out = pd.concat([out, remaining.sample(min(need, len(remaining)), random_state=seed)])
    return out.sample(frac=1, random_state=seed).reset_index(drop=True)

discovery_briefs = stratified_brief_sample(brief_objects, n=240, seed=42)

# this is what you send to discovery LLM
discovery_items = discovery_briefs[["brief_row_id", "description", "term_variants"]].rename(
    columns={"brief_row_id": "row_id"}
).to_dict("records")

print("brief objects:", len(brief_objects))
print("discovery items:", len(discovery_items))
print("example item keys:", discovery_items[0].keys() if discovery_items else "none")
````

````python
# 3) Discovery pass (many examples -> candidate classes)
DISCOVERY_SYSTEM = """
You are deriving an emergent taxonomy from successful search rows.

Goal:
1) discover brief types
2) discover query-strategy types used to satisfy them

Use only evidence in rows.
Return STRICT JSON:
{
  "brief_type_candidates":[
    {"label":"", "definition":"", "signals":[""], "example_row_ids":[""]}
  ],
  "query_strategy_candidates":[
    {"label":"", "definition":"", "moves":[""], "pattern_template":"", "example_row_ids":[""]}
  ],
  "notes":[]
}
"""

CHUNK = 5  # fewer because each item now contains multiple term variants
candidate_packets = []

total_chunks = (len(discovery_items) - 1) // CHUNK + 1

for i in range(0, 2, CHUNK): #len(discovery_items)
    chunk_id = i // CHUNK
    print(f"Processing chunk {chunk_id + 1} / {total_chunks}")
    chunk = discovery_items[i:i+CHUNK]  # already list[dict]
    out = ask_json(DISCOVERY_SYSTEM, {"chunk_id": chunk_id, "briefs": chunk})
    out["chunk_id"] = chunk_id
    candidate_packets.append(out)

len(candidate_packets)
````

````python
# 1) Larger sample + robust fq_values parsing
import ast, json, re, pandas as pd

sample_n = min(20000, len(good_collections))  # increase as needed
sample_df = good_collections.sample(n=sample_n, random_state=42).copy()

def parse_fq_values(x):
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    if isinstance(x, list):
        return [str(v).strip() for v in x if str(v).strip()]

    txt = str(x).strip()
    if not txt:
        return []

    for parser in (json.loads, ast.literal_eval):
        try:
            out = parser(txt)
            if isinstance(out, list):
                return [str(v).strip() for v in out if str(v).strip()]
        except Exception:
            pass

    # fallback for loosely formatted strings
    txt = txt.strip("[]")
    parts = [p.strip().strip('"').strip("'") for p in txt.split(",") if p.strip()]
    return parts

sample_df["fq_list"] = sample_df["fq_values"].map(parse_fq_values)
fq_tokens = sample_df["fq_list"].explode().dropna().astype(str).str.strip()
fq_tokens = fq_tokens[fq_tokens != ""]
````

````python
# 2) Inspect discovered filters + build generic NL conversion
def split_filter_token(tok):
    tok = tok.strip()
    if ":" not in tok:
        return tok, ""
    k, v = tok.split(":", 1)
    return k.strip(), v.strip()

parsed = fq_tokens.to_frame(name="filter_token")
parsed[["filter_key", "filter_value"]] = parsed["filter_token"].apply(
    lambda t: pd.Series(split_filter_token(t))
)

print("Unique filter tokens:", parsed["filter_token"].nunique())
display(parsed["filter_token"].value_counts().head(50))
display(parsed["filter_key"].value_counts().head(30))

def filter_token_to_nl(tok):
    k, v = split_filter_token(tok)
    vl = v.lower()

    if vl in ("true", "false"):
        return f"apply boolean filter: {k} = {vl}"
    if v.startswith("[") and " TO " in v:
        return f"apply range filter on {k}: {v}"
    if v.startswith("(") and v.endswith(")"):
        return f"apply set/logic filter on {k}: {v}"
    if v:
        return f"apply equality filter: {k} = {v}"
    return f"apply filter: {k}"

sample_df["filters_nl"] = sample_df["fq_list"].map(
    lambda arr: "; ".join(filter_token_to_nl(t) for t in arr)
)
````
